```python
import csv
f = open('age.csv','r',encoding = 'cp949')
data = csv.reader(f)
result = [ ]
for row in data :
        #빈 리스트 만들기
    if '신도림' in row [0] : #'신도림'이 포함된 행정구역 찾기
        for i in row[3:] : #0세부터 끝(100세 이상)까지 모든 연령에 대해 반복하기
            result.append(int(i)) #해당 연령의 인구수 리스트에 순서대로 저장하기
print(result)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

plt.figure(figsize = (10, 5), dpi = 300)
plt.plot(result)
plt.show()
```

```python
import csv
f = open('age.csv','r',encoding = 'cp949')
data = csv.reader(f)
result = []
name = input(' 인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해주세요. :')

for row in data :
        #빈 리스트 만들기
    if name in row [0] : #'신도림'이 포함된 행정구역 찾기
        for i in row[3:] : #0세부터 끝(100세 이상)까지 모든 연령에 대해 반복하기
            result.append(int(i.replace(',',''))) #해당 연령의 인구수 리스트에 순서대로 저장하기

print(result)            
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', family = 'HYPost')

plt.title(name +' 지역의 인구 구조')
plt.plot(result)
plt.show()
```
