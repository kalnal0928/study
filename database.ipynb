```python
%matplotlib inline
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Windows 환경에서 NanumGothic 폰트 경로 지정
font_path = r'C:\Windows\Fonts\NanumGothic.ttf'  # 또는 실제 폰트 파일 위치로 수정
if os.path.exists(font_path):
    fe = fm.FontEntry(fname=font_path, name='NanumGothic')
    fm.fontManager.ttflist.insert(0, fe)
    plt.rcParams.update({'font.size': 18, 'font.family': 'NanumGothic'})
else:
    print("NanumGothic.ttf 파일을 찾을 수 없습니다.")

# 폰트 확인
sys_font = fm.findSystemFonts()
print(f"sys_font number: {len(sys_font)}")
print(sys_font)

nanum_font = [f for f in sys_font if 'Nanum' in f]
print(f"nanum_font number: {len(nanum_font)}")
print(nanum_font)

pitcher = pd.read_csv('pitcher_stats_2017.csv')
batter = pd.read_csv('batter_stats_2017.csv')

print(pitcher.columns)
print(batter.columns)

print(pitcher.head())
print(batter.head())

# 예측할 대상인 '연봉'에 대한 정보
print(pitcher['연봉(2018)'].describe())
```
