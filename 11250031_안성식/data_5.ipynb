```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data :
    if row[-1] != '':
        month = row[0].split('-')[1]
#8월의 최고 기온 값을 저장할 aug 리스트 생성
#로 구분된 값 중 2번째 값을 month에 저장
    
        if month == '08': #8월달이라면
            aug.append(float(row[-1])) #aug 리스트에 최고 기온 값 추가
            
        if month == '01': #8월달이라면
            jan.append(float(row[-1])) #aug 리스트에 최고 기온 값 추가

plt.figure(dpi = 300)            
plt.hist(aug, bins=100, color='r')
plt.hist(jan, bins=100, color='b')
plt.legend()
plt.show()
```
