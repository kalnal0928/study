```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if row[-1] != '': #최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1])) #result 리스트에 최고 기온 값 추가
print(result)

plt.plot(result, 'r')
plt.show()
```

```python
# 두 번째 셀 수정
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []  # 초기화

for row in data :
    if row[-1] != '' : #최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '08' : #8월에 해당하는 값이라면
            result.append(float(row[-1])) #result 리스트에 최고 기온값 추가
            
plt.plot(result, 'hotpink')
plt.show() 
f.close()
```

```python
# 세 번째 셀 수정
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
max_temp = []  # 최고 기온
min_temp = []  # 최저 기온

for row in data :
    if row[-1] != '' and row[-2] != '':  # 최고기온과 최저기온 모두 존재할 때
        if row[0].split('-')[1] == '04' and row[0].split('-')[2] == '07':
            max_temp.append(float(row[-1]))  # 최고 기온
            min_temp.append(float(row[-2]))  # 최저 기온

plt.rc('font', family ='Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')
plt.plot(max_temp, 'red', label='최고 기온')
plt.plot(min_temp, 'blue', label='최저 기온')
plt.title('4월 7일 최고/최저 기온 변화')
plt.xlabel('연도')
plt.ylabel('기온(°C)')
plt.legend()
plt.grid(True)
plt.show() 

f.close()
```
