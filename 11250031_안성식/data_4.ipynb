```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if row[-1] != '': #    
        result.append(float(row[-1])) #result     
print(result)

plt.plot(result, 'r')
plt.show()
```

```python
#    
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []  # 

for row in data :
    if row[-1] != '' : #   
        if row[0].split('-')[1] == '08' : #8  
            result.append(float(row[-1])) #result    
            
plt.plot(result, 'hotpink')
plt.show() 
f.close()
```

```python
#    
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
max_temp = []  #  
min_temp = []  #  

for row in data :
    if row[-1] != '' and row[-2] != '':  #     
        if row[0].split('-')[1] == '04' and row[0].split('-')[2] == '07':
            max_temp.append(float(row[-1]))  #  
            min_temp.append(float(row[-2]))  #  

plt.rc('font', family ='Malgun Gothic')
plt.title('    ')
plt.plot(max_temp, 'red', label=' ')
plt.plot(min_temp, 'blue', label=' ')
plt.title('4 7 /  ')
plt.xlabel('')
plt.ylabel('(C)')
plt.legend()
plt.grid(True)
plt.show() 

f.close()
```
